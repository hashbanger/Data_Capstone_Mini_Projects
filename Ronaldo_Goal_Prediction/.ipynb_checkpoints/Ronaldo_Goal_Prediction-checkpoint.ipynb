## Reading the Data

```python
df = pd.read_csv('ronaldo.csv')
df.head()
```

```python
df = df.set_index(df['Unnamed: 0'] + 1)
df = df.drop(labels = 'Unnamed: 0', axis = 1)
```

Let's take a look at the missing values.

```python
plt.figure(figsize=(9,6))
sns.heatmap(df.isnull(), cbar= False)
plt.show()
```

So we can see that there's plenty of missing values. Especially in the columns **type_of_shot** and **type_of_combined_shot**

Since **is_goal** is our target variable, the missing values in that column are the values that we have to predict using our trained model so we will not drop or impute that column.

Firstly, we drop some of the duplicate columns at the end, the ones having ***'.1'*** as suffix. Secondly, we also drop columns which probably have no logical relation to goal like, **match_event_id**

```python
drop_feat = ['match_event_id', 'remaining_min.1', 'power_of_shot.1', 'knockout_match.1', 'remaining_sec.1', 'distance_of_shot.1'
             ,'match_id', 'team_id', 'shot_id_number', 'lat/lng']
```

```python
df = df.drop(drop_feat, axis = 1)
df.head()
```

### Missing Value Ratio   
In this approach we first calculate the density of missing values and fill them or drop them depending on the threshold of missing values percentage that we set.   
A logical step now here would be to drop the columns having missing values percentage above 20% and impute the ones having less.  

```python
df.isnull().sum()/len(df) * 100
```

```python
missing_pct = df.isnull().sum()/len(df) * 100
```

```python
drop_feat = []
for col in df.columns:
    if missing_pct[col] > 20 and col != 'is_goal':
        drop_feat.append(col)
```

```python
df = df.drop(drop_feat, axis= 1)
df.head()
```

## Remaining Minute and Remaining Seconds


Initially we would think that these two are obviously correlated variables as remaining minutes can also be denoted in remaining seconds. But since we have no additional knowledge of data we can't assume that.

```python
plt.figure(figsize=(8,5))
plt.scatter(df['remaining_min'], df['remaining_sec'])
plt.xlabel('remaining_min')
plt.ylabel('remaining_sec')
plt.show()
```

```python
df['remaining_sec'].max()
```

From the above two cells we can observe that the remaining minutes and remaining seconds are not linearly related and actually the remaining minutes on combining with remaining seconds gives us the final remaining time.

So we can create a new column called **remaining_time** which will contain the total remaining time in seconds.

```python
df['remaining_time'] = df['remaining_min'] * 60 + df['remaining_sec']
df.drop(['remaining_min', 'remaining_sec'],axis= 1, inplace = True)
```

```python
plt.figure(figsize=(16,5))
sns.distplot(df[df['remaining_time'].isna() == False]['remaining_time']/60, color = 'teal')
plt.xlabel('Remaining Time in Minutes')
plt.title('Distribution of Remining Time')
plt.show()
```

#### Imputing with mean

```python
df['remaining_time'] = df['remaining_time'].fillna(int(df['remaining_time'].mean())).astype('int')
```

### Home/Away

```python
df['home/away'].unique()
```

First of all, it's weird Manchester United is facing NBA teams! Let's take a moment to let that sink in .

From the **home/away** we can separate if the game was home or away and can also slice out the opponent.

Initially we impute the missing values with the mode

```python
df['home/away'] = df['home/away'].fillna(df['home/away'].value_counts().index[0])
```

```python
df['opponent'] = df['home/away'].apply(lambda x : x[-3:])
```

```python
away = np.zeros(len(df))
for i in range(len(df)):
    if '@' in df['home/away'].iloc[i]:
        away[i] = 1
```

```python
df['away'] = away.astype(int)
df.drop(['home/away'], axis = 1, inplace = True)
df.head()
```
